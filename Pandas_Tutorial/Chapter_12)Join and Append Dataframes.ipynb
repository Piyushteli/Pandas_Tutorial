# join

```python
import pandas as pd
var_1=pd.DataFrame({"A":[1,2,3,4,5],"B":[11,12,13,14,15]})
var_2=pd.DataFrame({"C":[21,22,23,24,25],"D":[31,32,33,34,35]})
var_1.join(var_2)
```

# join Parameters

We have mainly pass four parameters in 1)how="inner" 2)how="outer" 3)how="right" 4)how="left"

```python
import pandas as pd
var_1=pd.DataFrame({"A":[1,2,3,4,5],"B":[11,12,13,14,15]},index=[1,2,3,4,5])
var_2=pd.DataFrame({"C":[21,22],"D":[31,32]},index=[1,2])

var_2.join(var_1,how="inner")        #It acts as Intersection
```

```python
import pandas as pd
var_1=pd.DataFrame({"A":[1,2,3,4,5],"B":[11,12,13,14,15]},index=[1,2,3,4,5])
var_2=pd.DataFrame({"C":[21,22],"D":[31,32]},index=[1,2])

var_2.join(var_1,how="outer")           #It acts as Union
```

```python
import pandas as pd
var_1=pd.DataFrame({"A":[1,2,3,4,5],"B":[11,12,13,14,15]},index=[1,2,3,4,5])
var_2=pd.DataFrame({"C":[21,22],"D":[31,32]},index=[1,2])

var_2.join(var_1,how="right")
```

```python
import pandas as pd
var_1=pd.DataFrame({"A":[1,2,3,4,5],"B":[11,12,13,14,15]},index=[1,2,3,4,5])
var_2=pd.DataFrame({"C":[21,22],"D":[31,32]},index=[1,2])

var_2.join(var_1,how="left")
```

# To handle the value Error

```python
import pandas as pd
var_1=pd.DataFrame({"A":[1,2,3,4,5],"B":[11,12,13,14,15]},index=[1,2,3,4,5])
var_2=pd.DataFrame({"C":[21,22],"B":[31,32]},index=[1,2])

var_2.join(var_1,how="outer",lsuffix="_2",rsuffix="_1")
```

# Append

This function do not through an Error when the column name is same

```python
import pandas as pd
var_1=pd.DataFrame({"A":[1,2,3,4,5],"B":[11,12,13,14,15]})
var_2=pd.DataFrame({"C":[21,22],"D":[31,32]})

var_1.append(var_2)
```

```python
import pandas as pd
var_1=pd.DataFrame({"A":[1,2,3,4,5],"B":[11,12,13,14,15]})
var_2=pd.DataFrame({"C":[21,22],"D":[31,32]})

var_1.append(var_2,ignore_index=True)
```

```python
import pandas as pd
var_1=pd.DataFrame({"A":[1,2,3,4,5],"B":[11,12,13,14,15]})
var_2=pd.DataFrame({"C":[21,22],"B":[31,32]})

var_1.append(var_2)
```
