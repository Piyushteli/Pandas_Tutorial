# dropna and fillna

```python
import pandas as pd
df=pd.read_csv("C:\\Users\\ADMIN\\Downloads\\Bengaluru_House_Data.csv")
df
```

```python
df.dropna()
```

```python
df.dropna(axis=1)     #drop the data according to columns
```

# If particular whole row contain the missing data. Then drop this row using this command

```python
df.dropna(how="all")
```

To remove the whole (NaN)rows        

```python
df.dropna(how="any")
```

# Drop the values according to column

```python
df.dropna(subset="society")
```

# when the column has only one or two or more NaN value then we use thresh() function

```python
df.dropna(thresh=9)          #if the column contain 9 NaN values it will drop
```
