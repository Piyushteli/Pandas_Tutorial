# Merge

```python
import pandas as pd
var_1=pd.DataFrame({"A":[1,2,3,4,5],"B":[11,12,13,14,15]})
var_2=pd.DataFrame({"A":[1,2,3,4,5],"C":[21,22,23,24,25]})

pd.merge(var_1,var_2)
```

# Merge Parameters

```python
import pandas as pd
var_1=pd.DataFrame({"A":[1,2,3,4],"B":[11,12,13,14]})
var_2=pd.DataFrame({"A":[1,2,3,5],"C":[21,22,23,24]})

pd.merge(var_1,var_2,how="left")
```

```python
import pandas as pd
var_1=pd.DataFrame({"A":[1,2,3,4],"B":[11,12,13,14]})
var_2=pd.DataFrame({"A":[1,2,3,5],"C":[21,22,23,24]})

pd.merge(var_1,var_2,how="right")
```

```python
import pandas as pd
var_1=pd.DataFrame({"A":[1,2,3,4],"B":[11,12,13,14]})
var_2=pd.DataFrame({"A":[1,2,3,5],"C":[21,22,23,24]})

pd.merge(var_1,var_2,how="outer")
```

```python
import pandas as pd
var_1=pd.DataFrame({"A":[1,2,3,4],"B":[11,12,13,14]})
var_2=pd.DataFrame({"A":[1,2,3,5],"C":[21,22,23,24]})

pd.merge(var_1,var_2,how="outer",indicator=True)
```

# If there are the names commons in the dataset

```python
import pandas as pd
df_1=pd.DataFrame({"A":[1,2,3,4,5],"B":[11,12,13,14,15]})
df_2=pd.DataFrame({"A":[10,20,30,40,50],"B":[100,200,300,400,500]})

pd.merge(df_1,df_2,left_index=True,right_index=True)
```

```python
import pandas as pd
df_1=pd.DataFrame({"A":[1,2,3,4,5],"B":[11,12,13,14,15]})
df_2=pd.DataFrame({"A":[10,20,30,40,50],"B":[100,200,300,400,500]})

pd.merge(df_1,df_2,left_index=True,right_index=True,suffixes=("name","python"))
```

# concat Parameters

```python
import pandas as pd
df1=pd.DataFrame({"A":[1,2,3,4,5],"B":[11,12,13,14,15]})
df2=pd.DataFrame({"A":[1,2,3],"B":[21,22,23]})
    
pd.concat([df1,df2],axis=1,join="inner")
```

```python
import pandas as pd
df1=pd.DataFrame({"A":[1,2,3,4,5],"B":[11,12,13,14,15]})
df2=pd.DataFrame({"A":[1,2,3,4,5],"B":[21,22,23,24,25]})
    
pd.concat([df1,df2],axis=1,keys=["D1","D2"])
```
