# Pivot() Function

```python
import pandas as pd
df=pd.DataFrame({"days":[1,2,3,4,5,6,7],"st_name":["Ravi","Sam","Mayur","Sam","Sam","Mayur","Ravi"],"Eng":[18,13,16,20,15,17,19],"Math":[19,12,10,18,20,16,12]})
df
```

```python
df.pivot(index="days",columns="st_name")
```

# For to get the single sub data

```python
df.pivot(index="days",columns="st_name",values="Eng")
```

# Used of Aggregation Function

```python
import pandas as pd
df_1=pd.DataFrame({"days":[1,2,3,4,5,6,7],"st_name":["Ravi","Sam","Ravi","Sam","Sam","Ravi","Ravi"],"Eng":[18,13,16,20,15,17,19],"Math":[19,12,10,18,20,16,12]})

df_1.pivot_table(index="st_name",columns="days",
               aggfunc="sum")
```

```python
import pandas as pd
df_1=pd.DataFrame({"days":[1,2,3,4,5,6,7],"st_name":["Ravi","Sam","Ravi","Sam","Sam","Ravi","Ravi"],"Eng":[18,13,16,20,15,17,19],"Math":[19,12,10,18,20,16,12]})

df_1.pivot_table(index="st_name",columns="days",
               aggfunc="mean",margins="True")
```
