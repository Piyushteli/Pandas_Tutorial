# Melt() and Pivot() Function

This Both function are mainly used to reshape your python pandas file

```python
import pandas as pd
df=pd.DataFrame({"days":[1,2,3,4,5,6,7],"Eng":[18,13,16,20,15,17,19],"Math":[19,12,10,18,20,16,12]})
df
```

```python
pd.melt(df)
```

```python
pd.melt(df,id_vars=["days"])
```

```python
pd.melt(df,id_vars=["days"],var_name="Sub",value_name="Score")
```
